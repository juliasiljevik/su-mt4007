```python
import pandas as pd

# Ensure 'Antal' is of string type before applying .str accessor
data['Antal'] = data['Antal'].astype(str)

# Extract only numeric values from 'Antal'
data['Antal'] = data['Antal'].str.extract('(\d+)', expand=False)

# Convert to numeric type and handle errors (convert non-numeric values to NaN)
data['Antal'] = pd.to_numeric(data['Antal'], errors='coerce')

# Remove rows where 'Antal' is NaN (because they couldn't be converted to numbers)
data = data.dropna(subset=['Antal'])

# Group by 'Artnamn' and sum 'Antal'
species_counts = data.groupby('Artnamn')['Antal'].sum().reset_index()

# Sort by 'Antal' in descending order
species_counts = species_counts.sort_values(by='Antal', ascending=False)

# Display the most prevalent species
print("The most prevalent species are:")
species_counts.head(10) 
```

The data shows the most observed bird species, with Grönsiska leading at over 20,000 counts, followed by Sothöna (8,308) and Gräsand (7,167).

```python
import pandas as pd

# Group the data by smoking status and stroke occurrence
smoking_vs_stroke = data.groupby(['smoking_status', 'stroke']).size().unstack().fillna(0)

# Rename columns for clarity
smoking_vs_stroke.columns = ['No Stroke', 'Stroke']

# Calculate the total count per smoking status
smoking_vs_stroke['Total'] = smoking_vs_stroke['No Stroke'] + smoking_vs_stroke['Stroke']

# Calculate the percentage of stroke for each smoking category
smoking_vs_stroke['Percentage Stroke'] = (smoking_vs_stroke['Stroke'] / smoking_vs_stroke['Total']) * 100

# Format the 'Percentage Stroke' column to include the "%" sign
smoking_vs_stroke['Percentage Stroke'] = smoking_vs_stroke['Percentage Stroke'].apply(lambda x: f"{x:.2f}%")

# Display the updated table with percentages
smoking_vs_stroke
```

The dataset highlights the relationship between smoking status and stroke risk. The "formerly smoked" group has the highest stroke percentage (7.91%), followed by "smokes" (5.32%), "never smoked" (4.76%), and "unknown" (3.04%). This suggests past smoking may have a stronger long-term impact on stroke risk than current smoking. Despite being the largest group, "never smoked" shows a lower stroke percentage, reinforcing the link between smoking history and stroke. The "unknown" group’s low stroke percentage may reflect limited data or incomplete classification.

# Data Preparation

```python
import pandas as pd
import numpy as np

data = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/cell_phones_total.csv")
data.head(10)
```

```python
import pandas as pd
import numpy as np

# Load data
data = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/cell_phones_total.csv")

# Print initial column types and info
print("Initial Data Information:")
print(data.info())

# Function to convert values like 'k', 'M', 'B' to numeric
def convert_to_numeric(value):
    if isinstance(value, str):
        value = value.strip()  # Remove extra spaces
        multipliers = {'k': 1e3, 'M': 1e6, 'B': 1e9}
        for suffix, multiplier in multipliers.items():
            if value.endswith(suffix):
                return float(value.replace(suffix, '')) * multiplier
        try:
            return float(value)  # Try converting to float if not a k/M/B value
        except ValueError:
            return np.nan  # If it can't be converted, return NaN
    return value

# Show column types before conversion
print("\nColumn Types Before Conversion:")
print(data.dtypes)

# Apply the conversion function to all relevant columns (skip the first one which is iso-3)
data.iloc[:, 1:] = data.iloc[:, 1:].apply(lambda col: col.apply(convert_to_numeric))

# Show column types after conversion
print("\nColumn Types After Conversion:")
print(data.dtypes)

# Handle missing values: Forward fill and then fill remaining NaNs with 0
print("\nHandling Missing Values:")
print(f"Missing values before cleaning:\n{data.isnull().sum()}")
data.ffill(axis=0, inplace=True)  # Forward fill missing values (fill NaNs with the previous row's value)
data.fillna(0, inplace=True)  # Fill any remaining NaNs with 0
data = data.infer_objects()  # Ensures proper type inference after filling
print(f"\nMissing values after cleaning:\n{data.isnull().sum()}")
```

```python
# Sort by the year 2015 and display a sample of cleaned data
sorted_data = data.sort_values(by='2015', ascending=False)
print("\nCleaned Data Sample (sorted by 2015):")
sorted_data[['iso-3', '2015', '2016', '2017', '2018', '2019']].head()
```

 I handled missing values in two steps: forward-filling and filling remaining NaNs with zeros. Forward-filling propagates the most recent valid data to fill missing values, which is useful for time-series data where trends are often carried forward. Afterward, I filled any remaining NaNs with zeros to ensure the dataset was complete and ready for analysis. This approach avoids errors and ensures that the absence of data is treated as having no cellphones in those entries, maintaining dataset consistency and accuracy.
