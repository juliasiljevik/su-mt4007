```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import sqlite3 as sql
import matplotlib.pyplot as plt
import re
import os
```

```python
tennis_data= pd.read_csv("/Users/siljevik/tennis_data.csv")
tennis_data.head(4)
```

```python
# Filtrera för rader där FirstSetWinner är 1
first_set_winners = tennis_data[tennis_data['FirstSetWinner'] == 1].copy()

# Lägg till kolumnen
first_set_winners['FirstSetWinnerWonMatch'] = first_set_winners['FirstSetWinner'] == first_set_winners['Result ']

# Kontrollera distributionen
print("Sannolikheten att den som vinner första set också vinner matchen (endast för de som vann första set):")
print(first_set_winners['FirstSetWinnerWonMatch'].value_counts())

# Beräkna procentandelar
percentage = first_set_winners['FirstSetWinnerWonMatch'].value_counts(normalize=True) * 100

# Skriv ut resultaten
print("\nProcentandel för de som vann första set och också vann matchen:")
print(percentage)
```

```python
# Plotta cirkeldiagrammet
# Lägg till kolumnen
first_set_winners['FirstSetWinnerWonMatch'] = first_set_winners['FirstSetWinner'] == first_set_winners['Result ']

counts = first_set_winners['FirstSetWinnerWonMatch'].value_counts(normalize=True)
counts.plot(kind='pie', autopct='%1.1f%%', labels=['Vann matchen', 'Förlorade matchen'], colors=['lightblue', 'pink'])
plt.title('Resultat för de som vann första set och matchen:')
plt.ylabel('')
plt.show()
```

```python
from scipy.stats import chi2_contingency

# Skapa en korsningstabell
contingency_table = pd.crosstab(tennis_data['FirstSetWinner'], tennis_data['Result '])

# Utför Chi2-test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi2-statistik: {chi2}")
print(f"P-värde: {p}")

if p < 0.05:
    print("Det finns ett statistiskt signifikant samband mellan att vinna första set och matchen.")
else:
    print("Det finns inget statistiskt signifikant samband.")
```

# Jämför STOCKHOLM OPEN 2024,2014 och 2004

```python
# Filtrera för Stockholm Open
stockholm_open = tennis_data[(tennis_data['Tournament'] == 'Stockholm Open') &
                               (tennis_data['Year'].isin([2004, 2014, 2024]))]

# Kontrollera datan
#stockholm_open.head(10)

# Gruppanalys: Första set-vinnare och matchresultat
grouped = stockholm_open.groupby(['Year', 'FirstSetWinner'])['Result '].value_counts(normalize=True).unstack()

# Plotta staplat stapeldiagram
grouped.plot(kind='bar', stacked=True, figsize=(10, 6), color=['red', 'green'])
plt.title('Fördelning av första set och matchresultat (Stockholm Open)')
plt.ylabel('Procentandel')
plt.xlabel('År och första set (0=Förlorare, 1=Vinnare)')
plt.legend(['Förlorade matchen', 'Vann matchen'])
plt.show()
```

```python
# Skapa en kolumn för att indikera om första set-vinnaren vann matchen
stockholm_open['FirstSetWinnerWonMatch'] = stockholm_open['FirstSetWinner'] == stockholm_open['Result ']

# Gruppanalys: Procentandel av första set-vinnare som vann matchen
yearly_comparison = stockholm_open.groupby('Year')['FirstSetWinnerWonMatch'].mean() * 100

# Plotta stapeldiagram
yearly_comparison.plot(kind='bar', color=['blue', 'pink', 'green'], figsize=(8, 6))
plt.title('Procentandel av matcher där första set-vinnaren vann (Stockholm Open)')
plt.ylabel('Procentandel')
plt.xlabel('År')
plt.xticks(rotation=0)
plt.show()
```

```python
# Data för varje år
data_2004 = stockholm_open[stockholm_open['Year'] == 2004]['FirstSetWinnerWonMatch'].astype(int)
data_2014 = stockholm_open[stockholm_open['Year'] == 2014]['FirstSetWinnerWonMatch'].astype(int)
data_2024 = stockholm_open[stockholm_open['Year'] == 2024]['FirstSetWinnerWonMatch'].astype(int)

from scipy.stats import ttest_ind

# 2004 vs 2014
t_stat_2004_2014, p_val_2004_2014 = ttest_ind(data_2004, data_2014, nan_policy='omit')

# 2014 vs 2024
t_stat_2014_2024, p_val_2014_2024 = ttest_ind(data_2014, data_2024, nan_policy='omit')

# 2004 vs 2024
t_stat_2004_2024, p_val_2004_2024 = ttest_ind(data_2004, data_2024, nan_policy='omit')

# Skriv ut resultaten
print("T-test mellan 2004 och 2014: t-stat =", t_stat_2004_2014, ", p-värde =", p_val_2004_2014)
print("T-test mellan 2014 och 2024: t-stat =", t_stat_2014_2024, ", p-värde =", p_val_2014_2024)
print("T-test mellan 2004 och 2024: t-stat =", t_stat_2004_2024, ", p-värde =", p_val_2004_2024)
```

Det finns inget mönster i skillanden, skillnaden kan vi inte säga att den beror på något. 

```python
from scipy.stats import f_oneway

# ANOVA-test för alla tre åren
f_stat, p_val_anova = f_oneway(data_2004, data_2014, data_2024)

# Skriv ut resultatet
print("ANOVA-test: F-statistik =", f_stat, ", p-värde =", p_val_anova)
```

# Järmför STOCKHOLM OPEN 2024 mot PACE SALK OPEN 2024

```python
# Filtrera för Stockholm Open 2024
stockholm_open_2024 = tennis_data[(tennis_data['Tournament'] == 'Stockholm Open') &
                                    (tennis_data['Year'] == 2024)]

# Filtrera för Pace Salk Open 2024
pace_salk_open_2024 = tennis_data[(tennis_data['Tournament'] == 'Pace salk open') &
                                    (tennis_data['Year'] == 2024)]

# Stockholm Open 2024
stockholm_open_2024['FirstSetWinnerWonMatch'] = stockholm_open_2024['FirstSetWinner'] == stockholm_open_2024['Result ']

# Pace Salk Open 2024
pace_salk_open_2024['FirstSetWinnerWonMatch'] = pace_salk_open_2024['FirstSetWinner'] == pace_salk_open_2024['Result ']
```

```python
# Procentandel för Stockholm Open
stockholm_win_rate = stockholm_open_2024['FirstSetWinnerWonMatch'].mean() * 100

# Procentandel för Pace Salk Open
pace_salk_win_rate = pace_salk_open_2024['FirstSetWinnerWonMatch'].mean() * 100

print(f"Stockholm Open 2024: {stockholm_win_rate:.2f}% av första set-vinnarna vann matchen.")
print(f"Pace Salk Open 2024: {pace_salk_win_rate:.2f}% av första set-vinnarna vann matchen.")

import matplotlib.pyplot as plt

# Fördelningar
stockholm_counts = [stockholm_win_rate, 100 - stockholm_win_rate]
pace_salk_counts = [pace_salk_win_rate, 100 - pace_salk_win_rate]

# Etiketter
labels = ['Vann matchen', 'Förlorade matchen']

# Skapa figuren med två subplotter
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Stockholm Open 2024
axes[0].pie(stockholm_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['green', 'red'])
axes[0].set_title('Stockholm Open 2024')

# Pace Salk Open 2024
axes[1].pie(pace_salk_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['green', 'red'])
axes[1].set_title('Pace Salk Open 2024')

# Justera layout
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import ttest_ind

# Konvertera till numeriska värden (1 för True, 0 för False)
stockholm_results = stockholm_open_2024['FirstSetWinnerWonMatch'].astype(int)
pace_salk_results = pace_salk_open_2024['FirstSetWinnerWonMatch'].astype(int)

# Utför t-test
t_stat, p_val = ttest_ind(stockholm_results, pace_salk_results, nan_policy='omit')

# Skriv ut resultatet
print(f"T-statistik: {t_stat}")
print(f"P-värde: {p_val}")

# Tolka resultatet
if p_val < 0.05:
    print("Det finns en statistiskt signifikant skillnad mellan turneringarna.")
else:
    print("Ingen statistiskt signifikant skillnad mellan turneringarna.")
```

# Järmför STOCKHOLM OPEN 2024 och DALLAS OPEN 2024 

```python
# Filtrera för Stockholm Open 2024
stockholm_open_2024 = tennis_data[(tennis_data['Tournament'] == 'Stockholm Open') &
                                    (tennis_data['Year'] == 2024)]

# Filtrera för Dallas Open 2024
dallas_open_2024 = tennis_data[(tennis_data['Tournament'] == 'Dallas Open') &
                                    (tennis_data['Year'] == 2024)]

# Stockholm Open 2024
stockholm_open_2024['FirstSetWinnerWonMatch'] = stockholm_open_2024['FirstSetWinner'] == stockholm_open_2024['Result ']

# Dallas Open 2024 
dallas_open_2024['FirstSetWinnerWonMatch'] = dallas_open_2024['FirstSetWinner'] == dallas_open_2024['Result ']

```

```python
from scipy.stats import ttest_ind

# Konvertera True/False till 1/0 för t-test
stockholm_results = stockholm_open_2024['FirstSetWinnerWonMatch'].astype(int)
dallas_results = dallas_open_2024['FirstSetWinnerWonMatch'].astype(int)

# T-test
t_stat, p_val = ttest_ind(stockholm_results, dallas_results, nan_policy='omit')

print(f"T-statistik: {t_stat}")
print(f"P-värde: {p_val}")

if p_val < 0.05:
    print("Det finns en statistiskt signifikant skillnad mellan Stockholm Open och Dallas Open.")
else:
    print("Ingen statistiskt signifikant skillnad mellan Stockholm Open och Dallas Open.")
```

Finns inget mönster 


```python
# Fördelning för Stockholm Open
stockholm_counts = stockholm_results.value_counts(normalize=True) * 100

# Fördelning för Dallas Open
dallas_counts = dallas_results.value_counts(normalize=True) * 100

# Plotta cirkeldiagram
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Stockholm Open
stockholm_counts.plot(kind='pie', autopct='%1.1f%%', labels=['Förlorade matchen', 'Vann matchen'],
                      colors=['green', 'red'], ax=axes[0])
axes[0].set_title('Stockholm Open 2024')

# Dallas Open
dallas_counts.plot(kind='pie', autopct='%1.1f%%', labels=['Förlorade matchen', 'Vann matchen'],
                   colors=['green', 'red'], ax=axes[1])
axes[1].set_title('Dallas Open 2024')

plt.tight_layout()
plt.show()
```

# Alla Gram Slam tuneringar 

```python
Gramslam_data= pd.read_csv("/Users/siljevik/GramSlamFinalstat.csv")
Gramslam_data.head(11)
```

```python
# Första Set Vinnare vs Matchresultat
plt.figure(figsize=(10, 6))
sns.countplot(x='FirstSetWinner', hue='Result', data=Gramslam_data)
plt.title('Första Set Vinnare vs Matchresultat')
plt.xlabel('Första Set Vinnare')
plt.ylabel('Antal Matcher')
plt.legend(title='Match Resultat', labels=['Förlorare', 'Vinnare'])
plt.grid(True)
plt.show()
```

```python
# Filtrera matcher där FirstSetWinner också vann hela matchen
first_set_and_match_winners = Gramslam_data[(Gramslam_data['FirstSetWinner'] == 1) & (Gramslam_data['Result'] == 1)]

# Kontrollera om Rank Winner är bättre (lägre) än Rank Looser
better_ranking_count = (first_set_and_match_winners['Rank Winner'] < first_set_and_match_winners['Rank Looser']).sum()

# Beräkna andelen
total_count = len(first_set_and_match_winners)
percentage_better_ranking = (better_ranking_count / total_count) * 100 if total_count > 0 else 0

print(f"Av de som vann första set och matchen hade {percentage_better_ranking:.2f}% en bättre ranking än sin motståndare.")
```
