```python
import requests
import json

# Define the endpoint and parameters
data = "https://api.nobelprize.org/2.1/nobelPrizes"
params = {"nobelPrizeCategory": "phy", "format": "json"}

# Fetch data
response = requests.get(data, params=params)

# Ensure the data was fetched successfully
if response.status_code == 200:
    print("Data fetched successfully!")
    # Extract the JSON response
    physics_data = response.json()

    # Format the output to match the desired structure and filter by award year 2022
    formatted_data = []
    for prize in physics_data.get("nobelPrizes", []):
        if prize.get('awardYear') == '1925':  # Filtrera för att endast ta med priser från 2022
            formatted_data.append({
                'awardYear': prize.get('awardYear', ''),
                'category': prize.get('category', {}),
                'categoryFullName': prize.get('categoryFullName', {}),
                'dateAwarded': prize.get('dateAwarded', ''),
                'prizeAmount': prize.get('prizeAmount', 0),
                'prizeAmountAdjusted': prize.get('prizeAmountAdjusted', 0),
                'links': prize.get('links', {}),
                'laureates': prize.get('laureates', [])
            })

    # Pretty print the filtered data
    print(json.dumps(formatted_data, indent=2, ensure_ascii=False))
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

# Web Scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialize lists to hold the data
upcs = []
titles = []
prices = []
ratings = []

# Loop over the first 3 pages (Page 1 to 3)
for page in range(1, 4):
    # Build URL for each page
    url = f"https://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to retrieve page {page}. Skipping...")
        continue
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all the book items on the page
    books = soup.find_all('article', class_='product_pod')
    
    for book in books:
        # Extract the link to the individual book page
        book_link = book.h3.a['href']
        book_url = f"https://books.toscrape.com/catalogue/{book_link}"
        book_response = requests.get(book_url)
        
        if book_response.status_code != 200:
            print(f"Failed to retrieve book details for {book_link}. Skipping...")
            continue
        
        # Parse the book's individual page
        book_soup = BeautifulSoup(book_response.content, 'html.parser')
        
        # Extract the title
        title = book_soup.find('div', class_='col-sm-6 product_main').h1.text.strip()
        
        # Extract the price
        price = book_soup.find('p', class_='price_color').text.strip()
        
        # Extract the rating
        rating_class = book.find('p', class_='star-rating')['class'][1]  # Rating class (One, Two, etc.)
        rating_map = {
            "One": "One",
            "Two": "Two",
            "Three": "Three",
            "Four": "Four",
            "Five": "Five"
        }
        rating = rating_map.get(rating_class, "Unknown")
        
        # Extract the UPC (Universal Product Code)
        table = book_soup.find('table', class_='table table-striped')
        upc = table.find('th', text='UPC')  # Find the UPC row in the table
        if upc:
            upc = upc.find_next_sibling('td').text.strip()
        else:
            upc = "Unknown"  # Handle missing UPC
        
        # Append the extracted details to the respective lists
        upcs.append(upc)
        titles.append(title)
        prices.append(price)
        ratings.append(rating)
```
